```python
x=0
y=1
count=0
total=0
while True:
    i=input("Enter data :")
    if i=='':
        break
    count+=1
    i=float(i)
    total+=i
    if (i%2==0):
        x+=i
    else:
        y*=i
print("Total elements:",count,"Sum :",x,"Product",y,"Average",total/count)
   
```
